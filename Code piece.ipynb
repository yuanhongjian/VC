```python
import pandas as pd
```

```python
cab_data = pd.read_csv("Cab_Data.csv")
city = pd.read_csv("City.csv")
customer_id = pd.read_csv("Customer_ID.csv")
transaction_id = pd.read_csv("Transaction_ID.csv")
```

```python
merge_1 = pd.merge(cab_data, transaction_id, how = "inner", on = 'Transaction ID')
```

```python
merge_2 = pd.merge(merge_1, city, how = 'inner', on = 'City')
```

```python
final_df = pd.merge(merge_2, customer_id, how = "inner", on = 'Customer ID')
```
